```python
from collections import Counter
import math
import matplotlib.pyplot as plt
import random
import statistics

%matplotlib inline

class Coin:

    def __init__(self):
        self.side = random.choice(['heads', 'tails'])

class FlipSet:

    def __init__(self, n):
        self.flip_list = []
        self.total_number_of_flips = 2 ** n


    def flip_coin(self):
        """Flips a coin the amount of times instantiated in the FlipSet"""
        for n in range(self.total_number_of_flips):
            coin = Coin()
            self.flip_list.append(coin.side)
        #print(self.flip_list)
        #print(len(self.flip_list))
        ht_counter = Counter(self.flip_list)
        print("this is flip_coin, coin flipped {} times".format(self.total_number_of_flips))
        return ht_counter


    def find_diff_heads_tails(self):
        ht_counter = Counter(self.flip_list)
        #print(ht_counter)
        ht_diff = ht_counter['heads'] - ht_counter['tails']
        #print("ht_diff is {}".format(ht_diff))
        return ht_diff

    def find_ratio_heads_tails(self):
        ht_counter = Counter(self.flip_list)
        if ht_counter['tails'] == 0:
            ht_ratio = 0
        else:
            ht_ratio = ht_counter['heads'] / ht_counter['tails']
            #round it so it's not so much to store
            ht_ratio = round(ht_ratio, 3)
            #print("ht_counter is {}".format(ht_ratio))
        return ht_ratio
```
